```python
import random

options =( "rock", "paper", "scissors")
your_score = 0
computer_score = 0


computer = random.choice(options)
player = input("rock, paper or scissors? (or end the game(type yes)) ").lower()

while player != "yes":
    if  player == computer :
        print(computer + " vs " + player)
        print("Tie!")
        
    elif player == "rock":
        if computer == "scissors":
            print(computer + " vs " + player)
            print("Congratulations .You win!")
            your_score += 1
        elif computer == "paper":
            print(computer + " vs " + player)
            print("You lose!")
            computer_score += 1
        
    elif player == "paper":
        if computer == "rock":
            print(computer + " vs " + player)
            print("Congratulations .You win!")
            your_score += 1
        elif computer == "scissors":
            print(computer + " vs " + player)
            print("You lose!")
            computer_score += 1
    
    elif player == "scissors":
        if computer == "paper":
            print(computer + " vs " + player)
            print("Congratulations .You win!")
            your_score += 1
        elif computer == "rock":
            print(computer + " vs " + player)
            print("You lose!")
            computer_score += 1
    else:
        print("That's not a valid play. Check your spelling!")
 
    print("Your score: " + str(your_score))
    print("Computer score: " + str(computer_score))
    computer = random.choice(options)
    player = input("rock, paper or scissors? (or end the game(type yes)) ").lower()
    
print("Bye!")
print("Final score: " + str(your_score))
print("Computer score: " + str(computer_score))
```
