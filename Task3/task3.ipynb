```python
from tkinter import *
import random
import string

def generate_password(length, choice):
    password = ""

    if choice == 1:
        for i in range(int(length)):
            password += random.choice(string.ascii_lowercase)
    elif choice == 2:
        for i in range(int(length)):
            password += random.choice(string.ascii_letters)
    elif choice == 3:
        for i in range(int(length)):
            password += random.choice(string.ascii_letters + string.digits + string.punctuation)

    password_entry.delete(0, END)
    password_entry.insert(0, password)

    
def copy_password(password_entry):
    password = password_entry.get()
    root.clipboard_clear()
    root.clipboard_append(password)
    root.update()

root = Tk()
root.title("Password Generator")
root.geometry("1000x1000")
root.configure(bg="black")

photo = PhotoImage(file="pass.png")

desired_width = 100
desired_height = 100
x_scale = photo.width() // desired_width
y_scale = photo.height() // desired_height

photo = photo.subsample(x_scale, y_scale)

img = Label(image=photo, bg="navy")
label = Label(root, text="Password Generator", font=("Arial", 25, "bold"), bg="orange", fg="black")

label.pack(padx=10, pady=10)
img.pack(padx=10, pady=10)

choice = IntVar()
Font = ("Arial", 15, "bold")

weak_radio = Radiobutton(root, text="Weak", font=Font, bg="red", fg="black", value=1, variable=choice)
medium_radio = Radiobutton(root, text="Medium", font=Font, bg="yellow", fg="black", value=2, variable=choice)
strong_radio = Radiobutton(root, text="Strong", font=Font, bg="green", fg="black", value=3, variable=choice)

weak_radio.pack(padx=10, pady=10)
medium_radio.pack(padx=10, pady=10)
strong_radio.pack(padx=10, pady=10)

length_label = Label(root, text="Length", font=Font, bg="black", fg="white")
length_label.pack(padx=10, pady=10)

length_box = Spinbox(root, from_=8, to=32, font=Font, bg="black", fg="white", width=10)
length_box.pack(padx=10, pady=10)

generate_button = Button(root, text="Generate", font=Font, bg="black", fg="white", command=lambda: generate_password(length_box.get(), choice.get()))
generate_button.pack(padx=10, pady=10)

password_entry = Entry(root, font=Font, bg="black", fg="white")
password_entry.pack(padx=10, pady=10)

copy_button = Button(root, text="Copy", font=Font, bg="black", fg="white", command=lambda: copy_password(password_entry))
copy_button.pack(padx=10, pady=10)

root.mainloop()
```
