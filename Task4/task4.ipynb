```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

class App(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Contact Book")
        self.geometry("500x500")
        self.configure(bg="lightblue")
        self.createWidgets()
        
    
    def createWidgets(self):
        self.Label = tk.Label(self, text="Contact Book", bg="darkblue",fg = "white",font=("Arial", 20, "bold"))
        self.Label.grid(row=0, column=0, columnspan=2, sticky="ew", padx=10, pady=10)
        self.nameLabel = tk.Label(self, text="Name", bg="lightblue",fg ="white",font=("Arial", 10, "bold"))
        self.nameLabel.grid(row=1, column=0, sticky="ew", padx=10, pady=10)
        self.nameEntry = tk.Entry(self)
        self.nameEntry.grid(row=1, column=1, sticky="ew", padx=10, pady=10)
        self.phoneLabel = tk.Label(self, text="Phone", bg="lightblue",fg ="white",font=("Arial", 10, "bold"))
        self.phoneLabel.grid(row=2, column=0, sticky="ew", padx=10, pady=10)
        self.phoneEntry = tk.Entry(self)
        self.phoneEntry.grid(row=2, column=1, sticky="ew", padx=10, pady=10)
        self.emailLabel = tk.Label(self, text="Email", bg="lightblue",fg ="white",font=("Arial", 10, "bold"))
        self.emailLabel.grid(row=3, column=0, sticky="ew", padx=10, pady=10)
        self.emailEntry = tk.Entry(self)
        self.emailEntry.grid(row=3, column=1, sticky="ew", padx=10, pady=10)
        self.addressLabel = tk.Label(self, text="Address", bg="lightblue",fg ="white",font=("Arial", 10, "bold"))
        self.addressLabel.grid(row=4, column=0, sticky="ew", padx=10, pady=10)
        self.addressEntry = tk.Entry(self)
        self.addressEntry.grid(row=4, column=1, sticky="ew", padx=10, pady=10)
        self.addButton = tk.Button(self, text="Add", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command= self.addContact)
        self.addButton.grid(row=5, column=0, sticky="ew", padx=10, pady=10)
        self.deleteButton = tk.Button(self, text="Delete", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command = self.deleteContact)
        self.deleteButton.grid(row=5, column=1, sticky="ew", padx=10, pady=10)
        self.updateButton = tk.Button(self, text="Update", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command = self.updateContact)
        self.updateButton.grid(row=6, column=0, sticky="ew", padx=10, pady=10)
        self.searchButton = tk.Button(self, text="Search", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command = self.searchContact)
        self.searchButton.grid(row=6, column=1, sticky="ew", padx=10, pady=10)
        self.viewButton = tk.Button(self, text="View", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command = self.viewContact)
        self.viewButton.grid(row=7, column=0, sticky="ew", padx=10, pady=10)
        self.viewAllButton = tk.Button(self, text="View All", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command = self.viewAllContact)
        self.viewAllButton.grid(row=7, column=1, sticky="ew", padx=10, pady=10)
        self.clearButton = tk.Button(self, text="Clear", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command=self.clear)
        self.clearButton.grid(row=8, column=0, sticky="ew", padx=10, pady=10)
        self.exitButton = tk.Button(self, text="Exit", bg="darkblue",fg ="white",font=("Arial", 10, "bold"),command= self.exitButton)
        self.exitButton.grid(row=8, column=1, sticky="ew", padx=10, pady=10)
        
        
    def run(self):
        self.mainloop()
        
    def addContact(self):
        name = self.nameEntry.get()
        phone = self.phoneEntry.get()
        email = self.emailEntry.get()
        address = self.addressEntry.get()
        if name and phone and email and address:
            
            with open("contacts.txt", "a") as file:
                file.write(f"{name},{phone},{email},{address}\n")
            
            messagebox.showinfo("Success", "Contact added successfully.")
            self.clear()
        else:
            messagebox.showerror("Error", "All fields must be filled.")
        
    
    def deleteContact(self):
        name = self.nameEntry.get()
        if name:
            with open("contacts.txt", "r") as file:
                lines = file.readlines()
            with open("contacts.txt", "w") as file:
                for line in lines:
                    if line.split(",")[0] != name:
                        file.write(line)
            messagebox.showinfo("Success", "Contact deleted successfully.")
            self.clear()
        else:
            messagebox.showerror("Error", "Name field must be filled.")
    
    def updateContact(self):
        name = self.nameEntry.get()
        phone = self.phoneEntry.get()
        email = self.emailEntry.get()
        address = self.addressEntry.get()
        if name and phone and email and address:
            with open("contacts.txt", "r") as file:
                lines = file.readlines()
            with open("contacts.txt", "w") as file:
                for line in lines:
                    if line.split(",")[0] != name:
                        file.write(line)
                    else:
                        file.write(f"{name},{phone},{email},{address}\n")
            messagebox.showinfo("Success", "Contact updated successfully.")
            self.clear()
        else:
            messagebox.showerror("Error", "All fields must be filled.")
    
    def searchContact(self):
        name = self.nameEntry.get()
        if name:
            with open("contacts.txt", "r") as file:
                lines = file.readlines()
            for line in lines:
                if line.split(",")[0] == name:
                    messagebox.showinfo("Success", f"Name: {line.split(',')[0]}\nPhone: {line.split(',')[1]}\nEmail: {line.split(',')[2]}\nAddress: {line.split(',')[3]}")
                    break
            else:
                messagebox.showerror("Error", "Contact not found.")
        else:
            messagebox.showerror("Error", "Name field must be filled.")
    
    def viewContact(self):
        name = self.nameEntry.get()
        if name:
            with open("contacts.txt", "r") as file:
                lines = file.readlines()
            for line in lines:
                if line.split(",")[0] == name:
                    messagebox.showinfo("Success", f"Name: {line.split(',')[0]}\nPhone: {line.split(',')[1]}\nEmail: {line.split(',')[2]}\nAddress: {line.split(',')[3]}")
                    break
            else:
                messagebox.showerror("Error", "Contact not found.")
        else:
            messagebox.showerror("Error", "Name field must be filled.")
    
    def viewAllContact(self):
        with open("contacts.txt", "r") as file:
            lines = file.readlines()
        for line in lines:
            messagebox.showinfo("Success", f"Name: {line.split(',')[0]}\nPhone: {line.split(',')[1]}\nEmail: {line.split(',')[2]}\nAddress: {line.split(',')[3]}")
    
    def clear(self):
        self.nameEntry.delete(0, tk.END)
        self.phoneEntry.delete(0, tk.END)
        self.emailEntry.delete(0, tk.END)
        self.addressEntry.delete(0, tk.END)
    
    def exitButton(self):
        self.destroy()
    
app = App()
app.run()

```
