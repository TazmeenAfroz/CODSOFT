```python
from tkinter import *
from tkinter.font import Font

def add_task():
    new_task = new.get()  
    if new_task:
        my_list.insert(END, new_task)  
        new.delete(0, END) 
        save_list()


def delete_task():
    index = my_list.curselection() 
    if index:
        my_list.delete(index)
        save_list()
          
        
def update_task():
    index = my_list.curselection()  
    if index:
        updated_task = new.get()  
        if updated_task:
            my_list.delete(index)  
            my_list.insert(index, updated_task)
            save_list()
            
def save_list():
    with open("tasks.txt", "w") as f:
        tasks = my_list.get(0, END)  
        for task in tasks:
            if task.endswith("\n"):  
                f.write(task)  
            else:
                f.write(task + "\n")
                
def load_list():
    try:
        with open("tasks.txt", "r") as f:
            tasks = f.readlines()  
            for task in tasks:
                task = task.strip()
                my_list.insert(END, task)
    except:
        return
    
label = Tk()
label.geometry("700x700")
label.maxsize(1000, 1000)
label.title("To Do List")
label.config(bg="white")

list_label = Label(text="To Do List", bg="blue", fg="white", padx=20, pady=20, font=("Arial", 25, "bold"))
info = Label(text = "1.For adding first write in the entry box and then click on add task. \n2.For updating first write the updated task in the entry box and then select the task from the list and click on update task button\n3.For deleting first select the task from the list and then click on delete task button", bg="white", fg="blue", padx=20, pady=20, font=("Arial", 10, "italic"))
photo = PhotoImage(file="todo.png", height=200, width=300)
img = Label(image=photo, bg="white", padx=20, pady=20)
img.pack(side=TOP, fill=X, anchor='nw')

list_label.pack(side=TOP, anchor='nw', fill=X)
info.pack(side=TOP, fill=X, anchor='nw')
my_font = Font(family="Brush Script MT", size=15, weight="bold")
my_frame = Frame(label)
my_frame.pack(pady=10)

my_list = Listbox(my_frame, font=my_font, width=15, height=5, bg="light blue", bd=0, fg="blue", selectbackground="grey", activestyle="none")

my_list.pack(side=LEFT, fill=BOTH)


bar = Scrollbar(my_frame)
bar.pack(side=RIGHT, fill=BOTH)


my_list.config(yscrollcommand=bar.set)
bar.config(command=my_list.yview)


new = Entry(label, width=35, borderwidth=5)
new.pack(padx=10, pady=10)


add = Button(label, text="Add Task", padx=10, pady=10, command=add_task)
add.pack()

delete = Button(label, text="Delete Task", padx=10, pady=10, command=delete_task)
delete.pack()

update = Button(label, text="Update Task", padx=10, pady=10, command=update_task)
update.pack()
    
load_list()
    
label.mainloop()

```
